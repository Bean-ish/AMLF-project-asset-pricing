```python
import matplotlib.pyplot as plt

data1 = pd.DataFrame(data1).sort_values(by='x')
data2 = pd.DataFrame(data2).sort_values(by='x')
data3 = pd.DataFrame(data3).sort_values(by='x')
data4 = pd.DataFrame(data4).sort_values(by='x')
data5 = pd.DataFrame(data5).sort_values(by='x')
data6 = pd.DataFrame(data6).sort_values(by='x')
data7 = pd.DataFrame(data7).sort_values(by='x')
data8 = pd.DataFrame(data8).sort_values(by='x')
data9 = pd.DataFrame(data9).sort_values(by='x')
```

```python
plt.figure(figsize=(25, 10))
labels = ['Agriculture, forestry and fishing','Mining','Construction','Manufacturing',
        'Transportation and public utilities','Wholesale trade','Retail trade','Finance, insurance and real estate',
        'Services']

for i, df in enumerate([data1, data2, data3, data4, data5, data6, data7, data8, data9]):
    plt.plot(df['x'].tolist(), df['y'].tolist(), linestyle='-', label=labels[i])

# Add labels and title
plt.axvline(0, color='black', linestyle='--', linewidth=1.0)

plt.xlabel('dolvol value')
plt.ylabel('ALE value')
plt.title('Graph of ale value of feature dolvol for model 1')
plt.legend()

plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```
