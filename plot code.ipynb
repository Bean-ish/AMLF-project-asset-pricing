```python
import matplotlib.pyplot as plt
import numpy as np

# Sample R^2 values for each model
all_r2 = [0.75, 0.80, 0.85]
top100_r2 = [0.70, 0.75, 0.80]
bottom100_r2 = [0.65, 0.70, 0.75]

# Number of models
num_models = len(all_r2)

# Set the width of the bars
bar_width = 0.13

# Set the x locations for the groups
index = np.arange(num_models)

# Model names
model_names = ['Enet', 'RF', 'NN1']

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(index, all_r2, bar_width, label='All')
plt.bar(index + bar_width, top100_r2, bar_width, label='Top-100')
plt.bar(index + 2 * bar_width, bottom100_r2, bar_width, label='Bottom-100')

# Add labels, title, and legend
plt.ylabel('R^2 Value')
plt.title('R^2 Values for Different Models')
plt.xticks(index + bar_width, model_names)  # Set the x-axis labels to model names
plt.legend()

# y limits
plt.ylim(-0.2, 1.0)

# Draw a solid line for the x-axis
plt.axhline(0, color='black', linewidth=1)

# Add a light grey grid to the background
plt.grid(color='grey', linestyle='--', linewidth=0.5)

# Show plot
plt.show()
```
