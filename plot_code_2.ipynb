```python
import matplotlib.pyplot as plt
import numpy as np

# Sample R^2 values for each model
all_r2 = [-0.4814, 0.734, -3.21902, -10.272]
top100_r2 = [-7.007, 1.998, -5.4067, -3.12899]
bottom100_r2 = [-2.869, -0.62, -1.20888, -4.52820]

# Number of models
num_models = len(all_r2)

# Set the width of the bars
bar_width = 0.13

# Set the x locations for the groups
index = np.arange(num_models)

# Model names
model_names = ['Enet', 'RF', 'NN1', 'NN2']

# color
colors = ['midnightblue', 'lightseagreen', 'yellow']

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(index, all_r2, bar_width, label='All', color = colors[0])
plt.bar(index + bar_width, top100_r2, bar_width, label='Top-100', color = colors[1])
plt.bar(index + 2 * bar_width, bottom100_r2, bar_width, label='Bottom-100', color = colors[2])

# Add labels, title, and legend
plt.ylabel('R^2 Value', fontname='Times New Roman', fontsize=12)
plt.title('R^2-oos for Different Models', fontname='Times New Roman', fontsize=12)
plt.xticks(index + bar_width, model_names)  # Set the x-axis labels to model names
plt.legend()

# y limits
plt.ylim(-10.0, 8.0)

plt.yticks(np.arange(-10, 8, 0.5))

# Draw a solid line for the x-axis
plt.axhline(0, color='black', linewidth=.5)

# Add a light grey grid to the background
plt.grid(color='grey', linestyle='--', linewidth=0.5)

# remove edge
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.gca().spines['left'].set_linestyle('-')
plt.gca().spines['bottom'].set_visible(False)

plt.gca().spines['left'].set_linewidth(0.5)

# Show plot
plt.show()

```
