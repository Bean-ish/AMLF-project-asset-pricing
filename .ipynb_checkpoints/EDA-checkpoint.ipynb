## down-sampling

```python
# sort company record by number of rows, desc,
gpsize = gy_.size()
gpsize = pd.DataFrame(gpsize).sort_values(by=[0], ascending=False)
```

```python
# index of 500 companies that has the most records, to a list
permno = gpsize.iloc[:500].index.tolist()
```

```python
# select rows of certain companies (228 per company)
mask = data['permno'].isin(permno)
ds_data = data[mask]

# reset index
df = ds_data.reset_index(drop=True)
```

```python
df['return(t-1)'].isna().value_counts()
# assumption proven: only missing the first term 500 points, not so bad

df.isna().sum()
# the worst one: nincr has 2794 missing values
```
